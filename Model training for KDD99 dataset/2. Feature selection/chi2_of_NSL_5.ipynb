```python
import pandas as pd
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv('KDD99_SMOTE.csv')
```

```python
df.head()
```

```python
X = df.drop('label', axis=1)
y = df['label']
X.shape, y.shape
```

```python
# Choose the number of top features you want to select
k = 15  # You can change this value based on your requirements

# Initialize the SelectKBest object with the chi2 scoring function
chi2_selector = SelectKBest(chi2, k=k)

# Fit the selector to your data
X_new = chi2_selector.fit_transform(X, y)
```

```python
X_new.shape
```

```python
print(X_new)
```

```python
# Get the selected feature indices
selected_features_indices = chi2_selector.get_support(indices=True)
print(selected_features_indices)
```

```python
# Get the names of the selected features
selected_features = X.columns[selected_features_indices]
print(selected_features)
```

```python
# Create a new DataFrame with the selected features
selected_features_df = df[selected_features]
print(selected_features_df)
```

```python
selected_features_df.head()
```

```python
selected_features_df['labels'] = y
```

```python
selected_features_df.head()
```

```python
selected_features_df.to_csv('KDD99_chi2_15.csv', index = False)
```
