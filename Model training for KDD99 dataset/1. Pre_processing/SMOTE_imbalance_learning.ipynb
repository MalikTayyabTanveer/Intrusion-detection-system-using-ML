```python
import pandas as pd
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('KDD99_Preprocessed.csv')
df.head()
```

```python
X = df.drop('label', axis=1)
y = df['label']
```

```python
# Assuming the binary labels are in a column named 'label'
label_counts = df['label'].value_counts()
print(label_counts)
```

```python
smote = SMOTE()
```

```python
# Apply SMOTE to the training data only (to avoid data leakage)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
df_resampled = pd.concat([X_resampled, y_resampled], axis=1)
df_resampled.head()
```

```python
label_counts = df_resampled['label'].value_counts()
print(label_counts)
```

```python
df_resampled.to_csv('KDD99_SMOTE.csv', index = False)
```
